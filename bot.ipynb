```python
def complex_strategy(engine: Engine, instrument:str)->None:
    engine.signal_set.append(MultiTimeframeTrendFollowing(symbol=instrument, test=False))
    engine.execute_trade(mt5=True)
    engine.signal_set.clear()
```

```python
def show_market(engine: Engine, instrument:str)->None:
    signal = MultiTimeframeTrendFollowing(symbol=instrument, test=False)
    engine.signal_set.append(signal)
    engine.show_instrument_condition()
    if signal.condition_met:
        print("condition met")
    else:
        print("condition not met")
    # engine.execute_trade(mt5=True)
    engine.signal_set.clear()
```

```python
def compute_sma(history_price):
    f = history_price["close"].to_frame()
    f["SMA50"] = f["close"].rolling(50).mean()
    f["SMA100"] = f["close"].rolling(100).mean()
    f["SMA200"] = f["close"].rolling(200).mean()
    f.dropna(inplace=True)
    return f
```
